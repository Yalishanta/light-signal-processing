## Обрабатка изображения (получение фурье для "сигналов" трех цветов)

```python
#открытие картинки

image = io.imread("2.jpg")########################################################################################
print(plt.imshow(image))
print(image.shape)
image = cv2.resize(image, (1189, 109))
Xr= image[:,:,0][np.shape(image)[0]//2]
Xg= image[:,:,1][np.shape(image)[0]//2]
print(Xg)
Xb= image[:,:,2][np.shape(image)[0]//2]
Y = [a for a in range(0, 0+1*len(Xr))]
#графики картинки
plt.figure(figsize=(50, 20))
plt.plot(Y,Xr, linewidth= 5, color='r')
plt.plot(Y,Xg, linewidth= 5, color='g')
plt.plot(Y,Xb, linewidth= 5, color='b')
plt.tick_params(axis='both', which='major', labelsize=30)
plt.title('Cacl2_0,0001', fontsize=35)#################################################################################
plt.show()

win = 13
filt = np.ones(win)/win
res = np.convolve(Xb, filt, mode='same')#сглаживание
threshold = 25
indices = np.where(res < threshold)[0]
subarrays = np.split(res, indices)

B = []
for i in range(len(subarrays)):
    
    if len(subarrays[i]) >1:
        B.append(subarrays[i])

for i in range(len(B) ):        
    np.mean(B[i])
    np.max(B[i])
    if len(np.where(res == np.max(B[i]))[0])!=1:
        ind = np.where (res== np.max(B[i]))[0][-1]
    else: ind = np.where (res== np.max(B[i]))[0][0]    
    print('B-Lx%i='%i, ind)
    print('B-Ly%i='%i, np.max(B[i]))
    N1=B[i][0:np.where (B[i]== np.max(B[i]))[0][0]]#разрезал распределение попалам
    N2=B[i][np.where (B[i]== np.max(B[i]))[0][0]:]#разрезал распределение попалам

    x = np.max(B[i])/2#середина распределения
    index1 = np.abs(N1 - x).argmin() #число 1 оторые по середине
    index2 = np.abs(N2 - x).argmin() #число 2 оторые по середине

    nearest_value1 = N1[index1]#значение индекса среднего числа в первой половинке распределения
    nearest_value2 = N2[index2]#значение индекса среднего числа во второй половинке распределения

    L_dist = abs(np.where(B[i]==nearest_value1)[0][0] - np.where(B[i]==nearest_value2)[0][0])
    print('B-Ldist_%i=' %i,L_dist)
print('')
    
res = np.convolve(Xr, filt, mode='same')  
threshold = 25
indices = np.where(res < threshold)[0]
subarrays = np.split(res, indices)
subarrays    
R = []
for i in range(len(subarrays)):
    
    if len(subarrays[i]) >1:
        R.append(subarrays[i])
for i in range(len(R) ):        
    np.mean(R[i])
    np.max(R[i])
    if len(np.where(res == np.max(R[i]))[0])!=1:
        ind = np.where (res== np.max(R[i]))[0][-1]
    else: ind = np.where (res== np.max(R[i]))[0][0]    
    print('R-Lx%i='%i, ind)
    print('R-Ly%i='%i, np.max(R[i]))
    N1=R[i][0:np.where (R[i]== np.max(R[i]))[0][0]]#разрезал распределение попалам
    N2=R[i][np.where (R[i]== np.max(R[i]))[0][0]:]#разрезал распределение попалам

    x = np.max(R[i])/2#середина распределения
    index1 = np.abs(N1 - x).argmin() #число 1 оторые по середине
    index2 = np.abs(N2 - x).argmin() #число 2 оторые по середине

    nearest_value1 = N1[index1]#значение индекса среднего числа в первой половинке распределения
    nearest_value2 = N2[index2]#значение индекса среднего числа во второй половинке распределения

    L_dist = abs(np.where(R[i]==nearest_value1)[0][0] - np.where(R[i]==nearest_value2)[0][0])
    print('R-Ldist_%i=' %i,L_dist)
print('')
    
res = np.convolve(Xg, filt, mode='same')  
threshold = 25
indices = np.where(res < threshold)[0]
subarrays = np.split(res, indices)
subarrays    
G = []
for i in range(len(subarrays)):
    
    if len(subarrays[i]) >1:
        G.append(subarrays[i])
for i in range(len(G) ):        
    np.mean(G[i])
    np.max(G[i])    
    if len(np.where(res == np.max(G[i]))[0])!=1:
        ind = np.where (res== np.max(G[i]))[0][-1]
    else: ind = np.where (res== np.max(G[i]))[0][0]    
    print('G-Lx%i='%i, ind)
    print('G-Ly%i='%i, np.max(G[i]))
    N1=G[i][0:np.where (G[i]== np.max(G[i]))[0][0]]#разрезал распределение попалам
    N2=G[i][np.where (G[i]== np.max(G[i]))[0][0]:]#разрезал распределение попалам

    x = np.max(G[i])/2#середина распределения
    index1 = np.abs(N1 - x).argmin() #число 1 оторые по середине
    index2 = np.abs(N2 - x).argmin() #число 2 оторые по середине

    nearest_value1 = N1[index1]#значение индекса среднего числа в первой половинке распределения
    nearest_value2 = N2[index2]#значение индекса среднего числа во второй половинке распределения

    L_dist = abs(np.where(G[i]==nearest_value1)[0][0] - np.where(G[i]==nearest_value2)[0][0])
    print('G-Ldist_%i=' %i,L_dist)
    
print('')    
# Fourier transform 

N = 140
FTr = 2*np.abs(fft(Xr))/N
FTb = 2*np.abs(fft(Xb))/N
FTg = 2*np.abs(fft(Xg))/N
print('FTg_init:', FTr)
print('FTg_init:', FTb)
print('FTg_init:', FTg)
plt.figure(figsize=(10, 10))
plt.stem(np.delete(FTr, 0), use_line_collection=True, markerfmt = 'Dr', linefmt = '--') #удалил 0 элемент массива
plt.stem(np.delete(FTb, 0), use_line_collection=True, basefmt='1', markerfmt = 'Db', linefmt = '--') #удалил 0 элемент массива
plt.stem(np.delete(FTg, 0), use_line_collection=True, basefmt='1', markerfmt = 'Dg', linefmt = '--') #удалил 0 элемент массива


plt.xlim([-1, 35])
plt.xlabel('frequency')
plt.title('Fourier transform исходных данных Cacl2_0,0001')#################################################################
plt.grid()

N = 150
FTr = 2*np.abs(fft(np.convolve(Xr, filt, mode='same')))/N
FTb = 2*np.abs(fft(np.convolve(Xb, filt, mode='same')))/N
FTg = 2*np.abs(fft(np.convolve(Xg, filt, mode='same')))/N
print('FTr_smooth:', FTr)
print('FTb_smooth:', FTb)
print('FTg_smooth:', FTg)
#plt.figure(figsize=(10, 10))
#plt.stem(np.delete(FTr, 0), use_line_collection=True, markerfmt = 'Dr', linefmt = '--') #удалил 0 элемент массива
#plt.stem(np.delete(FTb, 0), use_line_collection=True, basefmt='1', markerfmt = 'Db', linefmt = '--') #удалил 0 элемент массива
#plt.stem(np.delete(FTg, 0), use_line_collection=True, basefmt='1', markerfmt = 'Dg', linefmt = '--') #удалил 0 элемент массива

#plt.xlim([-1, 35])
#plt.xlabel('frequency')
#plt.title('Fourier transform сглаженный Cacl2_0,0001_2')#############################################################################
#plt.grid()
```
